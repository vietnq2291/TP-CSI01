```python
import numpy as np

def test(cases, test_func, condition_func):
    
    n = len(cases)
    print("Testing on {} cases.".format(n))
    
    count = 0
    for i, case in enumerate(cases):
        try:
            passed = condition_func(test_func, case)
            if passed:
                print("- Test {} PASSED.".format(i+1))
                count += 1
            else:
                print("- Test {} FAILED: Wrong Answer.".format(i+1))
        except Exception as ex:
            print("An exception occured at test {}: {}.".format(i+1, ex))
            
    if count == n:
        print("CONGRATULATIONS! All test cases passed!")
    else:
        print("Some test failed. Maybe check some code?")
```

## Bài 2. Tính Biểu Thức

```python
def test2(func):
    
    test_cases = []
    
    class Node:
        def __init__(self, data):
            self.data = data
            self.left = None
            self.right = None

    # tree 1
    root = Node('+')
    root.left, root.right = Node(1), Node(1)
    test_cases.append({'inp': root, 'out': 2})
    
    # tree 2
    root = Node('+')
    root.left, root.right = Node('/'), Node(0)
    root.left.left, root.left.right = Node(1), Node(2)
    test_cases.append({'inp': root, 'out': .5})
            
    # tree 3
    root = Node('+')
    root.left, root.right = Node(3), Node('*')
    root.right.left, root.right.right = Node('+'), Node(2)
    root.right.left.left, root.right.left.right = Node(5), Node(9)
    test_cases.append({'inp': root, 'out': 31})
    
    # tree 4
    root = Node('-')
    root.left, root.right = Node('*'), Node('/')
    root.left.left, root.left.right = Node(2), Node(3)
    root.right.left, root.right.right = Node(20), Node(5)
    test_cases.append({'inp': root, 'out': 2})
    
    # tree 5
    root = Node('+')
    root.left, root.right = Node('*'), Node(7)
    root.left.left, root.left.right = Node('/'), Node(5)
    root.left.left.left, root.left.left.right = Node('+'), Node(2)
    root.left.left.left.left, root.left.left.left.right = Node(100), Node('^')
    root.left.left.left.right.left, root.left.left.left.right.right = Node(2), Node(3)
    test_cases.append({'inp': root, 'out': 277})
    
    # tree 6
    root = Node('+')
    root.left, root.right = Node('/'), Node(2)
    root.left.left, root.left.right = Node('^'), Node('*')
    root.left.left.left, root.left.left.right = Node(5), Node(2)
    root.left.right.left, root.left.right.right = Node(5), Node(3)
    test_cases.append({'inp': root, 'out': 3.666666666666667})
 
    def condition_func(func, case):
        return func(case['inp']) == case['out']
    
    test(test_cases, func, condition_func)
```
