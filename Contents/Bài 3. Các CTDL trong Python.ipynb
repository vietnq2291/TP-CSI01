## 1.1. List

**List**, hay một số ngôn ngữ lập trình khác gọi là mảng (array). *List* thường được dùng để lưu các phần tử cùng loại. Tuy nhiên, *list* cũng có thể lưu trữ các phần tử khác nhau về kiểu dữ liệu.  

Các đặc trưng của *list* trong Python:  

- Có thể thay đổi (mutable): Có thể thêm, xóa và thay đổi giá trị các phần tử trong *list*.
- Động (dynamic): *List* trong Python là một mảng động, có nghĩa là vùng nhớ dùng để lưu trữ một *list* có thể thay đổi, hỗ trợ trong việc thêm và xóa các phần tử trong *list* mà không cần quan tâm nhiều đến kích thước khi khởi tạo.
- Truy cập ngẫu nhiên (random access): Việc truy vấn một phần tử trong *list* dựa vào index có độ phức tạp O(1)

### Khởi Tạo

```python
list_1 = []                      # empty list
list_2 = ['a', 'b', 'c']         # list with pre-set elements
list_3 = [x for x in range(10)]  # list comprehension
list_4 = list(range(10))         # the list() constructor
list_5 = list("abcdef")          # the list() constructor to string

print(list_4)
print(list_5)
```

### Truy Vấn

```python
print(list_2)        # the whole list
print(list_3[5])     # get element at index
print(list_4[:5])    # slicing
print(list_4[5:10])  # slicing
```

## 2.1. Set

**Set** lưu trữ các phần tử mà không quan tâm đến thứ tự. Các phần tử trong *set* không được trùng nhau.  

*Set* hoạt động tốt hơn *list* trong việc:
- Kiểm tra một phần tử có nằm trong tập hợp hay không
- Xóa các phần tử trùng nhau
- Thực hiện các phép toán trên tập hợp (phép giao, phép hợp, phép hiệu)

Các đặc trưng của *set*: 

- Có thể thay đổi (mutable)
- Động (dynamic)
- *Set* là một tập hợp không có thứ tự nên không thể truy vấn các phần tử trong *set* dựa vào index.

### Khởi Tạo

```python
set_1 = set()                   # empty set
set_2 = {'a', 'b', 'c'}         # set with pre-set elements
set_3 = {x for x in range(10)}  # set comprehension
set_4 = set(range(10))          # the set() constructor
set_5 = set("abcdef")           # the set() constructor to string

print(set_3)
print(set_5)
```

### Truy Vấn

```python
print(set_2)  # the whole set
```

```python
# get length
len(set_4)
```

```python
# check membership
print(5 in set_4)
print(5 not in set_4)
```

```python
# loop through elements
for el in set_5:
    print(el, end=' ')
```

### Một Số Phương Thức

**Xóa phần tử**

**Kiểm tra tập hợp con**

```python
set_1 = set("abcdef")
set_2 = set("abc")
set_3 = set("xyz")

# check if a set is a subset of another
print("set_2 is subset of set_1: {}".format(set_2.issubset(set_1)))
print("set_3 is subset of set_1: {}".format(set_3.issubset(set_1)))
print()

# check if a set is a superset of another
print("set_1 is superset of set_2: {}".format(set_1.issuperset(set_2)))
print("set_1 is superset of set_3: {}".format(set_1.issuperset(set_3)))
```

**Sao chép**

```python
set_1 = set("abcdef")

# create another set in memory
set_copy = set_1.copy()

# compare to normal assignment
set_nocopy = set_1

# remove a value
set_1.remove('c')
print("copy():    {}".format(set_copy))
print("no copy(): {}".format(set_nocopy))
```

**Phép toán trên tập hợp**

```python
set_1 = set("abcdef")
set_2 = set("defghi")

print("Intersection: {}".format(set_1.intersection(set_2)))  # (phép giao)
print("Union: {}".format(set_1.union(set_2)))                # (phép hợp)
print("Difference: {}".format(set_1.difference(set_2)))      # (phép hiệu)
```

## 2.2. Dictionary

**Dictionary**, hay một số ngôn ngữ lập trình gọi là *map* hay *hashmap*, lưu trữ các phần tử dưới dạng *key: value* mà không quan tâm đến thứ tự của key. Các key trong *dictionary* không được trùng nhau.  

*Lưu ý*: Từ Python 3.7.0 trở lên, thứ tự các key trong *dictionary* được bảo đảm.  

*Dictionary* thường được dùng để truy vấn dữ liệu của một phần tử dựa trên key. Tương tự với từ điển trong thực tế, ta có thể tìm định nghĩa của một từ thông qua key là chính từ vựng đó.  

Các đặc trưng của *dictionary*: 

- Có thể thay đổi (mutable)
- Động (dynamic)
- Không thể truy vấn các phần tử trong *dictionary* dựa vào index mà phải dựa vào key

### Khởi Tạo

```python
dict_1 = {}                                     # empty dict
dict_2 = {'jack': 4098, 'joe': 4127}            # dict with pre-set elements
dict_3 = {x: x**2 for x in range(10)}           # dict comprehension
dict_4 = dict([('jack', 4098), ('joe', 4127)])  # the dict() constructor
dict_5 = dict(jack=4098, joe=4127)              # the dict() constructor
dict_6 = dict.fromkeys(['jack', 'joe'], 2021)   # fromkeys() method, same value for all keys
dict_7 = dict.fromkeys(['jack', 'joe'])         # fromkeys() method

print(dict_3)
print(dict_5)
print(dict_6)
print(dict_7)
```

```python
# get length
len(dict_5)
```

```python
# check membership of KEYS
print('jack' in dict_5)
print('jack' not in dict_5)
```

```python
# loop through elements

print("Keys:   ", end='')    # loop through keys
for key in dict_3:  # or dict_3.keys()
    print(key, end=' ')
print()

print("Values: ", end='')  # loop through values
for value in dict_3.values():
    print(value, end=' ')
print()
    
print("Items: ", end=' ')  # loop through items
for key, value in dict_3.items():
    print("{} : {} |".format(key, value), end=' ')
```

# 3. So Sánh Hai Nhóm CTDL

**Đặc trưng của CTDL có thứ tự:**
- Thứ tự các phần tử được đảm bảo
- Các phần tử có thể trùng lặp
- Việc truy vấn một phần tử dựa vào index có độ phức tạp là O(1)
- Việc truy vấn một phần tử dựa vào giá trị có độ phức tạp là O(n)
- Đối với *list*: Có thể sắp xếp các phần tử (từ nhỏ đến lớn hoặc ngược lại)

**Đặc trưng của CTDL không có thứ tự:**
- Thứ tự các phần tử không được đảm bảo
- Các phần tử không thể trùng lặp (đối với *set*), key không thể trùng lặp (đối với *dict*)
- Không thể truy vấn một phần tử dựa vào index
- Đối với *set*: Việc truy vấn một phần tử dựa vào giá trị có độ phức tạp là O(1)
- Đối với *dict*: Việc truy vấn một phần tử dựa vào key có độ phức tạp là O(1)
- Đối với *dict*: Việc truy vấn một phần tử dựa vào giá trị có độ phức tạp là O(n)
- Không thể sắp xếp các phần tử

## So Sánh Tốc Độ Truy Vấn

```python
# calculate execution time
def cal_time(func):
    start_time = time.time()
    result = func()
    real_time = time.time() - start_time
    return real_time, result
```

```python
# initialize data structures with ONE MILLION items

ONE_MILLION = 1000000
np.random.seed(2021)

np_rand_1 = np.random.rand(ONE_MILLION)
np_rand_2 = np.random.rand(ONE_MILLION)
choice_1 = random.choice(np_rand_1)
choice_2 = random.choice(np_rand_2)
rand_ind = np.random.randint(ONE_MILLION)

rand_list = np_rand_1.tolist()
rand_set = set(np_rand_1)
rand_dict = dict(zip(np_rand_1, np_rand_2))

del np_rand_1
del np_rand_2
```

**Truy vấn phần tử dựa vào index/key**

```python
print("List: {} s | Complexity O(1)".format(cal_time(lambda: rand_list[rand_ind])[0]))
print("Dict: {} s | Complexity O(1)".format(cal_time(lambda: rand_dict[choice_1])[0]))
```

*List* và *dict* đều có độ phức tạp là O(1) khi truy vấn dựa vào index/key.  
Điểm khác biệt là index của *list* phải là các số tuần tự 0, 1, 2,... còn key của *dict* không cần theo bất kỳ quy luật nào.

**Truy vấn phần tử dựa vào giá trị**

```python
print("List: {} s | Complexity O(n)".format(cal_time(lambda: rand_list.index(choice_1))[0]))
print("List: {} s | Complexity O(n)".format(cal_time(lambda: choice_1 in rand_list)[0]))
print("Set:  {} s | Complexity O(1)".format(cal_time(lambda: choice_1 in rand_set)[0]))
print("Dict: {} s | Complexity O(n)".format(cal_time(lambda: list(rand_dict.keys())[list(rand_dict.values()).index(choice_2)])[0]))
```

*Set* thuận tiện hơn *list* và *dict* khi cần truy vấn phần tử dựa vào giá trị.  
*Dict* trong Python rất không thuận tiện khi cần truy vấn phần tử theo giá trị. Mọi cách thực hiện đều phải qua bước chuyển đổi key và value thành *list*.

## 4.2. Đặc Trưng Của Dãy Số

**Yêu cầu**: Nhập vào một dãy số, trả về các giá trị: số lớn nhất, số nhỏ nhất và số trung bình cộng dưới dạng *tuple*.

```python
def get_summary(arr):
    max_num = max(arr)
    min_num = min(arr)
    mean_num = sum(arr)/len(arr)
    return (max_num, min_num, mean_num)


num_list = np.random.randint(0, 10000, 20).tolist()
print("List: {}".format(num_list))
print("List summary: {}".format(get_summary(num_list)))
```

## 4.3. Định Dạng Chuỗi

**Yêu cầu**: Nhập vào một chuỗi ký tự gồm các ký tự không in hoa (lower case), mỗi từ cách nhau bằng dấu ';'. Thực hiện các công việc sau:
- Đảo các từ theo thứ tự ngược lại
- In hoa chữ cái đầu tiên của mỗi từ
- Để duy nhất một khoảng trắng giữa các từ

**Ví dụ**:  
- Input: "school;technology;mindx"  
- Output: "Mindx Technology School"

**Gợi ý**: Sử dụng phương thức reverse() trên *list* để đảo thứ tự

```python
def format_str(inp_str):
    word_list = inp_str.split(';')
    word_list.reverse()
    result = ' '.join(word_list)
    result = result.title()
    return result


format_str("school;technology;mindx")
```

**Yêu cầu**: Cho danh sách mã số sinh viên đã quét thẻ vào ngày 1/1/2021 và ngày 2/1/2021, tìm:

- Input:
  + Ngày 1/1: 101, 107, 359, 107
  + Ngày 2/1: 994, 928, 788, 780, 928, 101
- Output:
  + Ngày 2/1: 994, 928, 788, 780, 101
  + Ngày 1/1 hoặc 2/1: 101, 107, 359, 994, 928, 788, 780
  + Chỉ ngày 1/1: 107, 359

```python
def get_student_lists(list_1, list_2):
    
    set_1 = set(list_1)
    set_2 = set(list_2)
    
    set_uni = set_1.union(set_2)
    set_only_1 = set_1.difference(set_2)
    
    print("Day 2:      {}".format(set_2))
    print("Day 1 or 2: {}".format(set_uni))
    print("Day 1 only: {}".format(set_only_1))
    
    
get_student_lists([101, 107, 359, 107], [994, 928, 788, 780, 928, 101])
```

## 4.5. Tần Suất

**Yêu cầu**: Cho một đoạn văn gồm nhiều từ và dấu câu, đếm tần suất xuất hiện của từng chữ cái trong đoạn văn đó (không phân biệt chữ hoa và thường)

**Ví dụ**:
- Input: "The cat is chasing the rat."
- Output: t=4 | h=3 | e=2 | c=2 | a=3 | i=2 | s=2 | n=1 | g=1 | r=1 |

**Gợi ý**: Sử dụng phương thức isalpha() trên *string* để kiểm tra một ký tự có nằm trong bảng chữ cái hay không

```python
# METHOD 1
def count_1(inp_str):
  
    result = {}
    for el in inp_str.lower():
        if el.isalpha():
            if el in result:
                result[el] += 1
            else:
                result[el] = 1
  
    for key, value in result.items():
        print("{}={} | ".format(key, value), end='')
        
        
count_1("The cat is chasing the rat.")
```

```python
# METHOD 2, more compact in code
def count_2(inp_str):
  
    result = {}
    for el in inp_str.lower():
        if el.isalpha():
            result[el] = result.get(el, 0) + 1
  
    for key, value in result.items():
        print("{}={} | ".format(key, value), end='')
        
        
count_2("The cat is chasing the rat.")
```
